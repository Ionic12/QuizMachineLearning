```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import RandomizedSearchCV

# import datasheet yang akan digunakan
df = pd.read_excel('data/dataset_quiz.xlsx')

# menampilkan data head atau data teratas sebanyak 100 baris
df.head(50)
```

```python
df.isnull().sum()

#pengecekan pada semua kolom apakah terdapat data yang benilai null
```

```python
# dilakukan drop dikarenakan untuk kolom tersebut tidak memiliki sebuah data
df = df.drop('id_anggota5', axis=1)
df.head() 
```

```python
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])

# label encoder digunakan untuk memberi label
```

```python
df = df.dropna()
# penghapusan baris yang memiliki nilai null sehingga ketika proses selanjutnya pemerikasaan untuk kolom yang bernilai null maka sudah tidak ada lagi
df.isnull().sum()

# pengecekan data pada setiap kolom apakah terdapat data yang null atau tidak
```

```python
print(df['keterlibatan_vltr'].value_counts())

# penghitungan jumlah label yaitu antara Ya dan Tidak pada single kolom
```

```python
X = df.iloc[:, 6:]
y = df.iloc[:, 5]
y.head()

# pengambilan sebagai label yang akan digunakan
```

```python
# training dan test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Model klasifikasi disimpan pada atribut dt.
dt = DecisionTreeClassifier()

# Menggabunkan data yang sudah dilakukan training data
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
# hasil dari test set accuray , dengan yang mengasilkan 1.0

acc = accuracy_score(y_test, y_pred)
print("Data Test accuracy: {:.2f}".format(acc))
print(f"Data Train accuracy: {acc}")
```

```python
# inisiasi objek baru MultinomialNB
multi = MultinomialNB()

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=50)

# Menggabunkan data yang sudah dilakukan training data
multi.fit(X_train, y_train)

# Prediksi dengan data training
y_train_pred = multi.predict(X_train)

# Evaluasi akurasi training
acc_train = accuracy_score(y_train, y_train_pred)

# Prediksi test data
y_test_pred = multi.predict(X_test)

# Evaluasi model dengan metric akurasi 
acc_test = accuracy_score(y_test, y_test_pred)

# Print hasil evaluasi
print(f'Hasil accuracy data train: {acc_train}')
print(f'Hasil accuracy data test: {acc_test}')
```

```python
# Inisiasi obyek RepeatedKFold

repeat = RepeatedKFold(n_splits=4, n_repeats=3, random_state=50)
```

```python
dt = DecisionTreeClassifier()

#set parameter dengan criterion : gini, entropy, log_loss dengan max depth rentang nilai 5 - 11
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11))
}

# Inisiasi obyek RandomizedSearchCV dengan isi yaitu decision tree, parameter dan cv
random = RandomizedSearchCV(dt, param_distributions=params, cv=repeat)
%timeit random.fit(X, y)

#mengambil nilai evaluasi dari X dan y
score = random.score(X,y)

#cetak hasilnya
print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```
